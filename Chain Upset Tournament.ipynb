```python
# This program calculates the different ways you can list a "chain" upset tournament of size 6.
```

```python
# function for generating an adjacency matrix that represents a "chain" upset tournament of size n
def chain_matrix(n):
    
    # fill with 0s
    matrix = [[0 for i in range(n)] for j in range(n)]
    for i in range(n):
        for j in range(n):
            # upset condition 
            if i == j + 1 or i < j - 1:
                matrix[i][j] = 1 
    return matrix

# function that divides the path list into sublists of length 6 and deletes sublists that contain repeat numbers 
def process(list):

    # divide the path list into sublists of length 6
    path_lists = [list[i * 6:(i + 1) * 6] for i in range(len(list)//6)] 
      
    # delete paths that don't reach every vertex
    for i in reversed(range(len(path_lists))):
        if len(set(path_lists[i])) != len(path_lists[i]):
            del path_lists[i]
            
    return path_lists
                                                      
# create adjacency matrix of size 6
matrix = chain_matrix(6)


# algorithm for finding paths 
list = []
for i in range(6):
    list.append(i + 1)
    
    for j in range(6):
        if matrix[i][j] == 1:
            if len(list) % 6 == 0:
                list.append(i + 1)
            list.append(j + 1)
           
            for k in range(6):
                if matrix[j][k] == 1:
                    if len(list) % 6 == 0:
                        list.append(i + 1)
                        list.append(j + 1)
                    list.append(k + 1)
                    
                    for m in range(6):
                        if matrix[k][m] == 1:
                            if len(list) % 6 == 0:
                                list.append(i + 1)
                                list.append(j + 1)
                                list.append(k + 1)
                            list.append(m + 1)
                            
                            for n in range(6):
                                if matrix[m][n] == 1:
                                    if len(list) % 6 == 0:
                                        list.append(i + 1)
                                        list.append(j + 1)
                                        list.append(k + 1)
                                        list.append(m + 1)
                                    list.append(n + 1)
                                    
                                    for p in range(6): 
                                        if matrix[n][p] == 1:
                                            if len(list) % 6 == 0:
                                                list.append(i + 1)
                                                list.append(j + 1)
                                                list.append(k + 1)
                                                list.append(m + 1)
                                                list.append(n + 1)
                                            list.append(p + 1)

# process the list                                            
paths = process(list)

# output results
print('The number of listings for a "chain" upset tournament of size 6 is ' + str(len(paths)))

for i in range(len(paths)):
    print(paths[i])
    





```
